```python
def calculate_price_per_box(group):
    # Mengambil nilai max dari cond_price_unit dalam setiap grup
    max_cond_price_unit = group['cond_price_unit'].max()

    for index, row in group.iterrows():
        if row['base_uom'] == 'PCS':
            # Jika base_uom == PCS, kalikan rate dengan max cond_price_unit
            group.at[index, 'price_per_box'] = row['rate'] * max_cond_price_unit
        elif row['base_uom'] == 'BOX':
            # Jika base_uom == BOX, gunakan rate tanpa perkalian
            group.at[index, 'price_per_box'] = row['rate']

    # Menghitung rata-rata price_per_box per material_id dan menambahkannya sebagai kolom average
    group['average'] = group['price_per_box'].mean()
    return group

new = df_price.groupby('material_id').apply(calculate_price_per_box)
```

```python
channel_mapping = {
    'GT': ['WS 2', 'Retail', 'Retail Large', 'Top WS', 'WS 1', 'Retail Mode TO', 'Retail Cosmetic',
           'WS Cosmetic', 'WS Snack', 'WS Beverage', 'Retail Small', 'RetailLarge cosmetic'],
    'MT': ['MTI-SM RO PDA', 'MTI-MM', 'MTKA Supermarket 2', 'MTI-SM RO MOBILE', 'MTKA Modern Wholesal',
           'MTKA Mini Market', 'MTI-TOP SM', 'MTKA Supermarket', 'MTI-SPECIAL OUTLET', 'MTI-SM',
           'MTKA Hypermarket', 'MTKA - Premium SPM', 'MTKA Drug Pharma', 'MTKA WholesaleOnline',
           'MTI-BABY MILK STORE', 'MTI-PREMIUM', 'MTKA Convenience Sto'],
    'Others': ['WAMIDA', 'Toko Bumbu', 'Apotek / Toko Obat', 'Office Building', 'Spc Baby', 'Restaurant',
               'Laundry', 'Prof./Hospital', 'Hotel', 'Horeca Telesales', 'Catering', 'Local Trading',
               'Industrial', 'Bakery']
}

def map_to_channel(customer_group):
    for channel, groups in channel_mapping.items():
        if customer_group in groups:
            return channel
    return 'Others'

df_cust['channel'] = df_cust['customer_group'].apply(map_to_channel)
```
